```python
# as usual we import the Modules that we need for the analysis
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("seaborn-whitegrid")
```

```python
# lets initalize a random numpy array
array = np.linspace(0,10,1000)
array.shape
```

```python
# initialize a figure object
# the most easiest one is probably a line diagram
fig = plt.figure() # here we can further decide on how many plots and the figure size
ax = plt.axes()
ax.plot(array,np.sin(array), color = "k") # we can change the color with the color
ax.plot(array, np.cos(array), color = "green", label = "cos(x)") # we can also add a label

######################################
# set axes options
ax.set_title("Sinus versus Cosinus")
ax.set_ylabel("Y-value")
ax.set_xlabel("X-value")

#####################################
# limit axes
ax.set_xlim(-1,11)
ax.set_ylim(-0.5, 0.5)

####################################
# axes can be set tight
plt.axis("tight")
plt.legend()
```

```python
array_scatter = np.linspace(0,10,50)

fig = plt.figure() # here we can further decide on how many plots and the figure size
ax = plt.axes()
plt.plot(array_scatter,
         np.sin(array_scatter),
         "o", 
         color = "k") # we can change the color with the color
plt.plot(array_scatter, 
         np.cos(array_scatter),
         "o", 
         color = "green", 
         label = "cos(x)") # we can also add a label
```

```python
# therefore we could also just do
fig = plt.figure
ax = plt.axes()
colors = np.random.rand(50)
sizes = 500* np.random.rand(50)
# this function can be also used to give different sizes to the different points
# this can act as additional dimension for the viewer
plt.scatter(array_scatter, np.sin(array_scatter), color = "green", s = 15, label = "Data1")
plt.scatter(array_scatter, 
            np.cos(array_scatter), 
            c = colors, 
            s = sizes, 
            cmap = "viridis", 
            label = "Data2")
plt.legend()
```

```python
# we can also visualize errors 
error = 0.5
plt.errorbar(array_scatter, np.sin(array_scatter), yerr = error)
# but error can be also of length of the sequence
y_err = np.random.rand(50)
plt.errorbar(array_scatter, np.cos(array_scatter), yerr = y_err)
# we can also change the style of the errorbar 
plt.errorbar(array_scatter, 
             array_scatter, 
             yerr = y_err, 
             elinewidth=3, 
             ecolor="red",
             fmt = 'o',
             capsize = 0)
```

```python
# Create a heatmap
# This example is directly from the matplotlib homepage
vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])


fig, ax = plt.subplots()
im = ax.imshow(harvest, cmap="twilight_shifted_r")

# Show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(farmers)), labels=farmers)
ax.set_yticks(np.arange(len(vegetables)), labels=vegetables)
# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
#plt.grid(None) # since grids are not that useful for heatmaps
plt.show()
```
