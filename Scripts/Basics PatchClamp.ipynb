```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# this is called a constant in python 
# PEP-8 says you are not allowed to change also ofc it is possible
df = pd.read_csv("../Data/Electrophysiology/IV/Processed_Files/iv_condition.csv")
df.head()
# here we check the first 5 rows of the table
```

```python
# how can we acess columns by name
# using lists allows us to select multiple columns
# this will not change the original dataframe
# it is just a view
# to change it I would need to store it in the variable df
print(df[["Sweep #","Ipeak[A]"]])
df
```

```python
# this is used to set specific styles using seaborn 
# first set(font_scale) increases the default font size by 1.2
sns.set(font_scale=1.2)
# this checks the style for seaborn
# there are various options available:
# you can find them: https://www.python-graph-gallery.com/104-seaborn-themes
sns.set_style("whitegrid")

# lets make a more beautiful plot
fig, ax = plt.subplots(figsize = (12,5), ncols = 2, nrows = 1)
sns.despine()
sns.lineplot(data = df , 
             x = "Sweep #", # select the x axis
             y = "Ipeak[A]", # select the y axis
             palette = "viridis", # this creates the colors that we would like to use for our plot
             hue = "condition", # here we can select different condition
             ax = ax[0], # this determines where to draw the plot
             legend = True, # this draws the little legend box based on your hue 
             errorbar=('se'), # this can be the metric of the errorbar, such as condidence interval, standard deviation or SEM
             err_style= "bars" # here we can decide either bands or bars
             )

sns.lineplot(data = df, 
             x = "Sweep #", 
             y = "Isus[A]", 
             palette = "viridis",
             hue = "condition", 
             ax = ax[1], 
             legend = False,
             errorbar=('se'),
             err_style = "bars"
             )

# make the appearance of the different tabs
# you could of course also write the axis labeling into a loop
ax[0].set_xlabel("Number of Sweeps")
ax[0].set_ylabel("Current in A")
ax[1].set_xlabel("Number of Sweeps")
ax[1].set_ylabel("Current in A")
# finally we set a title for our plot
plt.suptitle("IV Analysis")
```

```python
# have a look at the distribution of a single column
sns.histplot(df["Ipeak[A]"])
```

```python
# this plots checks correlation and distributino of all your columns in your dataset at once
sns.pairplot(df,hue = "condition")
```

```python
# how can we select columns based on specific features
print(df.columns)
# this can be trained again using list comprehensions
selected_columns_a = [i for i in df.columns if "[A]" in i]
selected_columns_a
```

```python
# let us draw a composition figure that hold all the columns
# you could try to put this into a function with the dataframe, and the column as parameter
fig = plt.figure(figsize = (15,15))
ax = plt.axes()

for index, column in enumerate(selected_columns_a):
    plt.subplot(3,3,index + 1)
    sns.lineplot(data = df, x = "Sweep #", y = column, hue = "condition")
    plt.xlabel("Current in A")
    plt.ylabel("Number of Sweeps")
    plt.title(column)

plt.suptitle("All IV parameters")    
plt.tight_layout()
```

```python
#slice the dataframe for the two columns
#.loc is used if you would like to name the column/index directly
# the usage of .loc is .loc[row_name: column_name]
# if you want to add multiple columns or rows use lists to select them, like shown below
df.loc[:,["Ipeak[A]","Ipp_bas[A]"]]
```

```python
# how would we select only the sweep between 5 and 15

wt_sweep_5_15 = df.loc[(df["Sweep #"]>5) & (df["Sweep #"] < 15) & (df["condition"] == "wt")]
wt_sweep_5_15
```

```python
# how can we change one element that is located in a cell
# we can select the specific cell
iv_change = df.copy(deep = True)
iv_change.iloc[1,1] = "hello"
iv_change
```

```python
# how would we set a whole columnn to a single value
# instead we could also add a list, but be aware the list must of equal size as the dataframe
df["amplifier"] = "HEKA"
df
```

```python
# groupby operation

grouped_df = df.groupby("Sweep #")["Ipeak[A]"].agg(["mean", "std"])
pd.DataFrame(grouped_df)
```

```python
# we can group this particular based on aggregation function
# you can imaging this as a reduced view
# we can add as many aggregation function as we want
# we could also create our own aggregation functions using lambda

grouped_df = df.groupby(["Sweep #", "condition"])[selected_columns_a].agg(["mean","std"])
pd.DataFrame(grouped_df)
```

```python
# last thing how can we make calculations
# of course we can also add other calculations like + - * as well as you own functions
df["normalized_ipeak"] = df["Ipeak[A]"]/df["Cslow[F]"]
normalized_table = df[selected_columns_a].copy(deep=True) # this creates a complet new copy of the dataframe in the reference
# unfortunately it is often not clear in pandas if the operation returns a view/reference or a copy 
# therefore when in doubt and you need to reuse you dataframe make a copy!
normalized_table = normalized_table.divide(df["Cslow[F]"], axis = 0)
normalized_table
```
