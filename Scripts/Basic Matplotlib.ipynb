# This is a module to show the drawing capabilities of python

```python
# as usual we import the Modules that we need for the analysis
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# lets initalize a random numpy array
array = np.linspace(0,10,100)
array.shape
```

```python
# initialize a figure object
# the most easiest one is probably a line diagram
fig = plt.figure() # here we can further decide on how many plots and the figure size
ax = plt.axes()
ax.plot(array,np.sin(array), color = "k") # we can change the color with the color
ax.plot(array, np.cos(array), color = "green", label = "cos(x)") # we can also add a label

######################################
# set axes options
ax.set_title("Sinus versus Cosinus")
ax.set_ylabel("Y-value")
ax.set_xlabel("X-value")

#####################################
# limit axes
ax.set_xlim(-1,11)
ax.set_ylim(-0.5, 0.5)

####################################
# axes can be set tight
plt.axis("tight")
plt.legend()
```

```python
fig = plt.figure() # here we can further decide on how many plots and the figure size
ax = plt.axes() # initialize the axis
plt.plot(array,
         np.sin(array),
         "o", 
         color = "k") # we can change the color with the color
plt.plot(array, 
         np.cos(array),
         "o", 
         color = "green", 
         label = "cos(x)") # we can also add a label
```

```python
# here you can see how we can change size and colors of the points based on weights such as in sizes or based on colors ranges
fig = plt.figure
ax = plt.axes()
colors = np.random.rand(100)
sizes = 500* np.random.rand(100)
# this function can be also used to give different sizes to the different points
# this can act as additional dimension for the viewer
plt.scatter(array, np.sin(array), color = "green", s = 15, label = "Data1")
plt.scatter(array, 
            np.cos(array), 
            c = colors, 
            s = sizes, 
            cmap = "viridis", 
            label = "Data2")
plt.legend()
```

```python
# in addition we can also 
fig = plt.figure
ax = plt.axes()
# this function can be also used to give different sizes to the different points
# this can act as additional dimension for the viewer
plt.scatter(array, np.sin(array), color = "green", s = 15, label = "Data1")
plt.scatter(array, # x-data 
            np.cos(array), #y-data
            c = colors,  # color range
            s = sizes, # size ranges
            cmap = "viridis",  # thsi describes the colormap used
            label = "Data2") # here the label name is setted for the legend
ax.legend(loc = "upper right", frameon = True, shadow = True, borderpad = 1) # loc lower center
```

```python
# we can also visualize errors 
error = 0.5
plt.errorbar(array, np.sin(array), yerr = error) # plot a sinus curve with an errorbar
# but error can be also of length of the sequence
y_err = np.random.rand(100)
plt.errorbar(array, np.cos(array), yerr = y_err) # plot a cosinus cure with an errobar
# we can also change the style of the errorbar 
plt.errorbar(array, 
             array, 
             yerr = y_err, 
             elinewidth=3, 
             ecolor="red",
             fmt = 'o',
             capsize = 0)
```

```python
# important for statistics, if you have a 1d array you can just easily get
# the histogram
# lets create a new array with. more data
# Excercise Create: one array with random/uniform distribution and three arrays that are normally distributed
# check the mean, std, median of each of the arrays
# then draw a histogram of the arrays to visualize
data = np.random.rand(1000) # randomly choosen
data_normal_1 = np.random.normal(3,2,1000) # choosen from normal distribution
data_normal_2 = np.random.normal(-1,2,1000) # mean, standard variation
data_normal_3 = np.random.normal(-1,7,1000)
```

```python
#create the figure
fig = plt.figure(figsize = (10,5))
ax = plt.axes()
# create a first subplot
plt.subplot(1,2,1)
plt.hist(data, color = "green", alpha = 0.5)
plt.title("Uniform distribution")
# create a second subplot
plt.subplot(1,2,2)
plt.hist(data_normal_1, bins = 30, color = "red", alpha = 0.5)
plt.hist(data_normal_2, bins = 30, color = "green", alpha = 0.5)
plt.hist(data_normal_3, bins = 30, color = "blue", alpha = 0.5)
plt.title("Normal Distribution")
```

```python
# two dimensional histograms
fig = plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
plt.hist2d(data_normal_1, data_normal_2, bins = 30, cmap = "viridis")
plt.colorbar(label = "Number of Data per Bin")
plt.subplot(1,2,2)
plt.hexbin(data_normal_2, data_normal_3, gridsize = 30, cmap = "viridis")
cb = plt.colorbar(label = "Number of Data per Bin")
plt.show() # this one can be called in a notebooks but is not need
# in a ide or other enviroments it is necessary to call this otherwise
# plot will not be shown
```

```python
# Create a heatmap
# This example is directly from the matplotlib homepage
vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])


fig, ax = plt.subplots()
im = ax.imshow(harvest, cmap="twilight_shifted_r")

# Show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(farmers)), labels=farmers)
ax.set_yticks(np.arange(len(vegetables)), labels=vegetables)
# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
#plt.grid(None) # since grids are not that useful for heatmaps
plt.show()
```

```python
# lets use a higher level plotting library
# Seaborn
# https://seaborn.pydata.org/examples/index.html
# install first
# this is just a scatterplot
import seaborn as sns

fig,ax = plt.subplots(figsize = (5,5))
sns.despine()
sns.scatterplot(x = data_normal_1, y = data_normal_2, linewidth = 0, alpha = 0.5)
plt.xlabel("Data 1")
plt.ylabel("Data 2")
plt.show()
```

```python
# this is a regression plot combining the fit of the best line with a scatterplot
fig,ax = plt.subplots(figsize = (5,5))
sns.despine()
sns.regplot(x = data_normal_1, y = data_normal_2)
plt.xlabel("Data 1")
plt.ylabel("Data 2")
plt.show()
```

```python
# first this
data_box = np.random.normal(2,5,50)
wt = np.array(["wt" for _ in range(25)])
ko = np.array(["ko" for _ in range(25)])
#than this
#male = np.array(["m" for _ in range(12)] + ["f" for _ in range(13)] )
#female = np.array(["f" for _ in range(13)] + ["m" for _ in range(12)])
condition = np.concatenate([wt, ko])
#group = np.concatenate([male, female])
```
