```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# this is called a constant in python 
# PEP-8 says you are not allowed to change also ofc it is possible
df = pd.read_csv("../Data/Electrophysiology/IV/Processed_Files/iv_condition.csv")
df.head()
# here we check the first 5 rows of the table
```

```python
# this creates an information table of your dataframe
# holding information about the type of the column
# about missing values
# about the shape of the df 
df.info()
```

```python
# this gives you basic descriptive statistics 
df.describe()
```

```python
# how can we acess columns by name
# using lists allows us to select multiple columns
# this will not change the original dataframe
# it is just a view
# to change it I would need to store it in the variable df
print(df[["Sweep #","Ipeak[A]"]])
df
```

```python
# this is used to set specific styles using seaborn 
# first set(font_scale) increases the default font size by 1.2
sns.set(font_scale=1.2)
# this checks the style for seaborn
# there are various options available:
# you can find them: https://www.python-graph-gallery.com/104-seaborn-themes
sns.set_style("whitegrid")

# lets make a more beautiful plot
fig, ax = plt.subplots(figsize = (12,5), ncols = 2, nrows = 1)
sns.despine()
sns.lineplot(data = df , 
             x = "Sweep #", # select the x axis
             y = "Ipeak[A]", # select the y axis
             palette = "viridis", # this creates the colors that we would like to use for our plot
             hue = "condition", # here we can select different condition
             ax = ax[0], # this determines where to draw the plot
             legend = True, # this draws the little legend box based on your hue 
             errorbar=('se'), # this can be the metric of the errorbar, such as condidence interval, standard deviation or SEM
             err_style= "bars" # here we can decide either bands or bars
             )

sns.lineplot(data = df, 
             x = "Sweep #", 
             y = "Isus[A]", 
             palette = "viridis",
             hue = "condition", 
             ax = ax[1], 
             legend = False,
             errorbar=('se'),
             err_style = "bars"
             )

# make the appearance of the different tabs
# you could of course also write the axis labeling into a loop
ax[0].set_xlabel("Number of Sweeps")
ax[0].set_ylabel("Current in A")
ax[1].set_xlabel("Number of Sweeps")
ax[1].set_ylabel("Current in A")
# finally we set a title for our plot
plt.suptitle("IV Analysis")
```
